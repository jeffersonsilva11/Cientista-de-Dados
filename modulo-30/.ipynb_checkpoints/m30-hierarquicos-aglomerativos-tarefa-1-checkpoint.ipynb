# Mtodos hierrquicos de agrupamento - Tarefa

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
import scipy.cluster.hierarchy as shc
```

## 1) Repetindo o que fizemos em aula

Primeiramente vamos repetir o que fizemos em aula com a base de pinguins.

1. Selecione as variveis quantitativas da base de pinguins
2. Elimine valores faltantes
3. Padronize as variveis para mdia zero e desvio padro 1
4. Faa um agrupamento hierrquico dividindo essa base em 3 grupos
5. Faa o Dendrograma

```python
# Carregando a base de pinguins
df = sns.load_dataset('penguins')
```

```python
# 1. Selecionando as variveis quantitativas
quantitative_vars = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']
df_quant = df[quantitative_vars]
```

```python
# 2. Eliminando valores faltantes
df_quant = df_quant.dropna()
```

```python
# 3. Padronizando as variveis
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_quant)
```

```python
# 4. Agrupamento hierrquico
cluster = AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage='ward')
cluster_id = cluster.fit_predict(df_scaled)
df_quant['Cluster'] = cluster_id
```

```python
# 5. Dendrograma
plt.figure(figsize=(10, 7))
plt.title("Dendrograma de Pinguins")
dend = shc.dendrogram(shc.linkage(df_scaled, method='ward'))
plt.show()
```

## 2) Avalie os grupos

Descreva os grupos:

- Quantos pinguins de cada espcie h em cada grupo?
- Descreva as variveis quantitativas de cada grupo com as tcnicas que voc preferir.

```python
# Selecionando as variveis quantitativas e removendo valores faltantes
df_quant = df[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']].dropna()
df = df.loc[df_quant.index]  # Atualizando o DataFrame original para refletir a remoo de valores faltantes
```

```python
# Padronizando as variveis
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_quant)
```

```python
# Realizando um agrupamento hierrquico
cluster = AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage='ward')
cluster_id = cluster.fit_predict(df_scaled)
df_quant['Cluster'] = cluster_id
```

```python
# Adicionando a coluna Cluster ao DataFrame
df['Cluster'] = cluster_id
```

```python
# Contagem dos pinguins por espcie
contagem_especies_por_cluster = df.groupby('Cluster')['species'].value_counts()
print(contagem_especies_por_cluster)
```

```python
medias_por_cluster = df.groupby('Cluster')[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']].mean()
print(medias_por_cluster)
```

O Dendrogarma sugere que podemem haver grupos interessantes se solicitarmos mais que 3 grupos. Se quisermos testar o agrupamento com mais de 3 grupos, quantos grupos o Dendrograma te sugere? (Dica: no faa mais que 6 grupos).

#### Ao olhar para o dendrograma anterior, podemos perceber que h nveis de corte que resultariam em mais de trs grupos. Podemos tentar um corte que forme 5 grupos para fins de anlise.

```python
# Realizando o agrupamento hierrquico com 5 grupos
```

```python
cluster = AgglomerativeClustering(n_clusters=5, metric='euclidean', linkage='ward')
df['Cluster'] = cluster.fit_predict(df[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']])
```

```python
# Analisando os agrupamentos para saber quantos pinguins de cada espcia h em cada grupo
contagem_especies_por_cluster = df.groupby('Cluster')['species'].value_counts()
print(contagem_especies_por_cluster)
```

```python
# Analisando os agrupamentos para saber quantos machos e fmeas h em cada grupo
contagem_especies_sexo_por_cluster = df.groupby(['Cluster', 'species'])['sex'].value_counts()
print(contagem_especies_sexo_por_cluster)
```

# Anlise:
    
Os resultados mostram que o agrupamento capturou alguns padres:<br><br>

- **Padres de Espcie:** Os pinguins Gentoo tendem a ser agrupados juntos (Clusters 2, 3, e 4), enquanto os Adelie e Chinstrap so muitas vezes agrupados juntos (Clusters 0 e 1). Isso pode indicar caractersticas semelhantes entre os pinguins Gentoo que os distinguem dos outros dois.
<br><br>
- **Padres de Sexo:**  H certa segregao baseada no sexo em alguns clusters, como visto no Cluster 1 (Adelie femininos) e Cluster 4 (Gentoo masculinos). Isso sugere que os pinguins masculinos e femininos de certas espcies podem ter caractersticas distintas que influenciam o agrupamento.

Em resumo, o agrupamento parece ter capturado padres naturais com base na espcie e no sexo dos pinguins. A distino dos pinguins Gentoo nos clusters, bem como a segregao baseada no sexo em alguns clusters, sugere que esses dois fatores desempenham um papel significativo na diferenciao dos pinguins.
