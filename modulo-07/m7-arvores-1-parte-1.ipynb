#### 2) Vamos montar um metadados

1. Crie um dataframe com os nomes de cada variável e o tipo de dados de cada variável.
2. Adicione uma coluna nesse *dataframe* chamada "qtd_categorias" e coloque nela o número de categorias correspondente de cada variável. 
    Dica: 
        1. inicie uma lista vazia
        2. faça um for ao longo dos nomes das variáveis, 
        3. conte o número de categorias dessa variável 
        4. acumule essa informação de 3. na lista que você criou em 1. 
        5. No final, essa lista pode ser atribuída à nossa variável.
3. Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}.

````python
# Crie o dataframe de metadados
metadados = pd.DataFrame({
    'varname': df_demo.columns,
    'dtype': df_demo.dtypes.values
})

# Crie a coluna 'qtd_categorias'
qtd_categorias = []
for var in metadados['varname']:
    qtd_categorias.append(df_demo[var].nunique())
metadados['qtd_categorias'] = qtd_categorias

metadados
````

````python
# Identifique as variáveis categóricas que não são binárias
categorical_vars = metadados.loc[(metadados['dtype'] == 'object') & (metadados['qtd_categorias'] > 2), 'varname']

# Crie variáveis dummy para essas variáveis
df_demo_dummies = pd.get_dummies(df_demo, columns=categorical_vars)

df_demo_dummies.head()
````

#### 3) Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}. Crie um *dataframe* apenas com as variáveis apropriadas para entrada no scikitlearn - elimine as variáveis tipo *str*, mantendo apenas suas versões *dummy*.

````python
# Primeiro, identificamos todas as colunas de tipo 'object'
cols_to_dummy = df_demo.select_dtypes(include='object').columns

# Criamos um novo DataFrame com as variáveis dummy
df_dummies = pd.get_dummies(df_demo, columns=cols_to_dummy)

df_dummies.head()
````

#### 4) Qual variável é mais poderosa?

Considere as variáveis ```possui_email``` e ```posse_de_veiculo```. Faça uma tabela cruzada entre elas e responda qual delas te parece mais poderosa para prever a probabilidade de ```mau = 1```?

````python
# Crie uma tabela cruzada para 'possui_email' e 'mau'
email_mau_crosstab = pd.crosstab(df_dummies['possui_email'], df_dummies['mau'], normalize='index')

# Crie tabelas cruzadas para 'posse_de_veiculo_N'/'posse_de_veiculo_Y' e 'mau'
veiculoN_mau_crosstab = pd.crosstab(df_dummies['posse_de_veiculo_N'], df_dummies['mau'], normalize='index')
veiculoY_mau_crosstab = pd.crosstab(df_dummies['posse_de_veiculo_Y'], df_dummies['mau'], normalize='index')

print(email_mau_crosstab)
print(veiculoN_mau_crosstab)
print(veiculoY_mau_crosstab)
````
