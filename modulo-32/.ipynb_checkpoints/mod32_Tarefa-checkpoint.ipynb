# Probabilidade - Tarefa

```python
import numpy as np
from scipy import stats
```

1. Uma mquina embala rao para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatrio, com distribuio Normal, com mdia 15kg, mas com desvio padro 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produo tenha mais que 15,2 kg.

```python
# Definindo a mdia e o desvio padro
mu = 15.0
sigma = 0.1

# Obtenhendo a probabilidade de que um pacote tenha peso menor ou igual a 15,2 kg
prob_menor_ou_igual_15_2 = stats.norm.cdf(15.2, mu, sigma)

# Probabilidade de que um pacote tenha mais que 15,2 kg
prob_maior_15_2 = 1 - prob_menor_ou_igual_15_2

print(f"A probabilidade de um pacote ter mais que 15,2 kg  {prob_maior_15_2:.4f}")
```

2. Da mquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

```python
# Definindo a mdia e o desvio padro
mu = 15.0
sigma = 0.1

# Obtenhendo o percentil de ordem 95%
percentil_95 = stats.norm.ppf(0.95, mu, sigma)

print(f"O percentil de ordem 95% para o peso dos pacotes  {percentil_95:.4f} kg")
```

3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na clula abaixo:

```python
# Definindo os valores em Y
Y = np.linspace(-6, 6, 200)

# Calcule a probabilidade de Y1 ser menor que cada valor em Y usando a CDF da distribuio normal padro
prob_Y1 = stats.norm.cdf(Y, 0, 1)

# Calcule a probabilidade de Y2 ser menor que cada valor em Y usando a CDF da distribuio t-student com 5 graus de liberdade
prob_Y2 = stats.t.cdf(Y, 5)

# Se desejar, voc pode imprimir algumas das probabilidades para verificar
for y, p1, p2 in zip(Y[::40], prob_Y1[::40], prob_Y2[::40]):  # Pega a cada 40 valores para no imprimir demais
    print(f"Para Y = {y:.2f}, P(Y1 < Y) = {p1:.4f}, P(Y2 < Y) = {p2:.4f}")

# Agora, prob_Y1 e prob_Y2 contm as probabilidades correspondentes para cada valor em Y
```

4. Faa uma visualizao grfica dos dados do item anterior.

```python
import matplotlib.pyplot as plt

# Plote as probabilidades
plt.figure(figsize=(10, 6))

# Grfico para a varivel Y1 com distribuio Normal
plt.plot(Y, prob_Y1, label='Normal', color='blue')

# Grfico para a varivel Y2 com distribuio t-student
plt.plot(Y, prob_Y2, label='t(5)', color='red')

# Detalhes do grfico
plt.title('CDF das variveis aleatrias Y1 e Y2')
plt.xlabel('Valor de Y')
plt.ylabel('Probabilidade')
plt.legend()
plt.grid(True)
plt.show()
```

5. 3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, o percentil (a inversa da *CDF*) para cada um dos valores do objeto *P* abaixo:

```python
Y=np.linspace(0, 1, 20)
# Seu cdigo comea aqui

from scipy.stats import norm, t

P = np.linspace(0, 1, 20)

# Para a distribuio Normal
quantiles_Y1 = norm.ppf(P)

# Para a distribuio t-student com 5 graus de liberdade
quantiles_Y2 = t.ppf(P, df=5)

for p, q1, q2 in zip(P, quantiles_Y1, quantiles_Y2):
    print(f"Para P = {p:.2f}, Y1: {q1:.4f}, Y2: {q2:.4f}")
```

6. Faa um grfico com os dados do item anterior.

```python
plt.figure(figsize=(10, 6))

# Plotar os dados
plt.plot(P, quantiles_Y1, label='Normal', color='blue')
plt.plot(P, quantiles_Y2, label='t(5)', color='red')

# Adicionar legendas e ttulos
plt.title('Quantis das distribuies Y1 e Y2')
plt.xlabel('Probabilidade')
plt.ylabel('Valor do Quantil')
plt.legend()
plt.grid(True)

plt.show()
```
