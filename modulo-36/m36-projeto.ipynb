### 4.4 Segmentação Geoespacial dos Dados

Para uma análise geoespacial mais detalhada, os dados podem ser divididos em células ou "bins" geográficos. Essa técnica permite agregar informações em áreas específicas, o que pode ser útil para identificar padrões ou tendências locais.

A seguir, é implementada uma função que segmenta os dados em bins geográficos baseados em uma resolução definida:

```python
def create_geo_bins(df, lat_res, lon_res):
    """
    Divide os dados em bins geoespaciais com base em uma resolução fornecida.
    
    Args:
    - df: DataFrame contendo colunas 'latitude' e 'longitude'
    - lat_res: resolução da latitude (e.g., 0.01 para células de 1km de altura aproximadamente)
    - lon_res: resolução da longitude (e.g., 0.01 para células de 1km de largura aproximadamente)
    
    Retorna:
    - DataFrame com uma nova coluna 'geo_bin' contendo os bins geoespaciais.
    """
    
    df['lat_bin'] = (df['latitude'] // lat_res).astype(int)
    df['lon_bin'] = (df['longitude'] // lon_res).astype(int)
    df['geo_bin'] = list(zip(df['lat_bin'], df['lon_bin']))
    
    return df

# Configuração dos bins
LAT_RESOLUTION = 0.01  # Ajuste conforme necessário
LON_RESOLUTION = 0.01  # Ajuste conforme necessário

df_airbnb = create_geo_bins(df_airbnb, LAT_RESOLUTION, LON_RESOLUTION)
```

### 4.5 Agregação por Células Geográficas

Após segmentar os dados em células geográficas, é útil agregar métricas específicas por célula para obter insights sobre diferentes regiões. Esta agregação fornece uma visão geral dos preços, número de avaliações e noites mínimas em cada célula, permitindo identificar áreas que possuem características específicas ou que se destacam de alguma maneira.

A seguir, os dados são agrupados por célula geográfica e agregados para calcular a média, mediana e desvio padrão das seguintes métricas:

- Preço
- Número de avaliações

```python
grouped_data = df_airbnb.groupby('geo_bin').agg({
    'price': ['mean', 'median', 'std'],
    'number_of_reviews': ['mean', 'median', 'std'],
    'minimum_nights': ['mean', 'median', 'std']  # Substitua 'nights' pelo nome real da coluna
})
```

### 4.6 Análise de Correlação por Célula Geográfica

Uma análise de correlação pode revelar as relações entre as diferentes variáveis em cada célula geográfica. Este passo é crucial para identificar padrões ou tendências em áreas específicas.

Aqui, calculamos a correlação entre:

- Preço e Número de Avaliações
- Preço e Noites Mínimas

Os resultados de correlação são, então, agregados para cada célula geográfica, produzindo as seguintes métricas:

- Média das Correlações
- Mediana das Correlações
- Desvio padrão das Correlações

Este resumo estatístico pode ajudar a identificar células geográficas onde as correlações são particularmente fortes ou fracas, ou onde existe uma grande variação nas correlações entre diferentes listagens.

```python
# Calcular as correlações para cada grupo 'geo_bin' e armazenar em uma lista
correlation_data = []

for group, data in df_airbnb.groupby('geo_bin'):
    corr_matrix = data[['price', 'number_of_reviews', 'minimum_nights']].corr()
    correlation_mean = corr_matrix.loc['price', ['number_of_reviews', 'minimum_nights']].mean()
    correlation_median = corr_matrix.loc['price', ['number_of_reviews', 'minimum_nights']].median()
    correlation_std = corr_matrix.loc['price', ['number_of_reviews', 'minimum_nights']].std()
    
    correlation_data.append([group, correlation_mean, correlation_median, correlation_std])

# Criar um DataFrame a partir da lista de correlações
correlation_df = pd.DataFrame(correlation_data, columns=['geo_bin', 'correlation_mean', 'correlation_median', 'correlation_std'])

# Resultado
print(correlation_df)
```

### 4.7 Análise dos Dados de Correlação Geográfica

```python
# Configurando o gráfico
plt.figure(figsize=(10, 6))

# Scatter plot
sns.scatterplot(x='correlation_mean', y='correlation_std', data=correlation_df, alpha=0.6)

# Título e rótulos dos eixos
plt.title('Média da Correlação vs. Desvio Padrão da Correlação')
plt.xlabel('Média da Correlação com o Preço')
plt.ylabel('Desvio Padrão da Correlação com o Preço')
plt.tight_layout()

# Exibir gráfico
plt.show()
```

### 4.9 Classificação de Avaliações
Para abordar o problema de classificação, a variável alvo foi definida da seguinte forma: se o número de avaliações é maior que zero, a variável alvo recebe o valor 1, caso contrário, recebe o valor 0.

Em seguida, foram selecionados os recursos relevantes para o modelo, incluindo 'room_type', 'minimum_nights', 'neighbourhood' e 'price'. O modelo utilizado foi o RandomForestClassifier.

A validação cruzada com 5 folds foi aplicada para avaliar o desempenho do modelo, utilizando a métrica de 'accuracy' para problemas de classificação. As pontuações de validação cruzada foram calculadas e a média das pontuações foi determinada como uma medida de desempenho.

```python
from sklearn.preprocessing import LabelEncoder

# Defina a variável alvo
df_airbnb['target_variable'] = np.where(df_airbnb['number_of_reviews'] > 0, 1, 0)

# Separe os recursos e a variável alvo
X = df_airbnb[['room_type', 'minimum_nights', 'neighbourhood', 'price']]
y = df_airbnb['target_variable']

# Codificação one-hot para 'room_type'
X = pd.get_dummies(X, columns=['room_type'], prefix=['room_type'])

# Codificação one-hot para 'neighbourhood'
X = pd.get_dummies(X, columns=['neighbourhood'], prefix=['neighbourhood'])

# Modelo (substitua pelo modelo de sua escolha)
model = RandomForestClassifier()

# Realize a validação cruzada com 5 folds
scores = cross_val_score(model, X, y, cv=5, scoring='accuracy')  # Use a métrica 'accuracy' para problemas de classificação

# Exiba as pontuações de validação cruzada
print("Pontuações de validação cruzada:", scores)

# Calcule a média das pontuações
mean_score = np.mean(scores)
print("Média das pontuações:", mean_score)
```

### 5.1 Treinamento de Modelo com Otimização de Hiperparâmetros
Nesta seção, realizamos o treinamento de um modelo de classificação mais avançado, o GradientBoostingClassifier, e otimizamos seus hiperparâmetros utilizando busca em grade.

```python
# Codificando one-hot para 'room_type'
X = pd.get_dummies(X, columns=['room_type'], prefix=['room_type'])

# Codificando one-hot para 'neighbourhood'
X = pd.get_dummies(X, columns=['neighbourhood'], prefix=['neighbourhood'])

# Escolhendo um modelo mais avançado
from sklearn.ensemble import GradientBoostingClassifier
model = GradientBoostingClassifier()

# Definindo hiperparâmetros para busca em grade
param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 4, 5]
}

# Realizando a pesquisa em grade com validação cruzada estratificada
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Inicializando a barra de progresso
progress_bar = tqdm(total=len(param_grid['n_estimators']) * len(param_grid['learning_rate']) * len(param_grid['max_depth']), 
                    desc="Grid Search Progress")

results = []

for n_estimators in param_grid['n_estimators']:
    for learning_rate in param_grid['learning_rate']:
        for max_depth in param_grid['max_depth']:
            model.set_params(n_estimators=n_estimators, learning_rate=learning_rate, max_depth=max_depth)
            scores = cross_val_score(model, X, y, cv=cv, scoring='accuracy')
            mean_score = np.mean(scores)
            results.append({
                'n_estimators': n_estimators,
                'learning_rate': learning_rate,
                'max_depth': max_depth,
                'mean_score': mean_score
            })
            progress_bar.update(1)  # Atualizando a barra de progresso

progress_bar.close()

# Encontrando a melhor combinação de hiperparâmetros
best_result = max(results, key=lambda x: x['mean_score'])
best_params = {
    'n_estimators': best_result['n_estimators'],
    'learning_rate': best_result['learning_rate'],
    'max_depth': best_result['max_depth']
}
best_score = best_result['mean_score']

print("Melhor combinação de hiperparâmetros:", best_params)
print("Melhor pontuação média de validação cruzada:", best_score)
```

### 5.2 Avaliação da Importância das Features
Aqui, ajustamos o modelo com os melhores hiperparâmetros encontrados anteriormente e avaliamos a importância das features no modelo.

```python
# Ajusta ndo o modelo com os melhores hiperparâmetros
best_model = GradientBoostingClassifier(**best_params)
best_model.fit(X, y)

# Avaliando a importância das features
feature_importances = best_model.feature_importances_
feature_names = X.columns

# Plotando a importância das features
N = 4  # por exemplo
sorted_idx = feature_importances.argsort()[-N:][::-1]
plt.figure(figsize=(10, 6))
sns.barplot(x=feature_importances[sorted_idx], y=feature_names[sorted_idx])
plt.title('Importância das Features no Modelo')
plt.xlabel('Importância')
plt.ylabel('Features')
plt.show()
```
