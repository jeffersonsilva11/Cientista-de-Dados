```python
# Importando as bibliotecas necessárias

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
sns.set_theme()
```

```python
# Carregando os dataframes para cada mês

# Março
sinasc_mar = pd.read_csv('./input/SINASC_RO_2019_MAR.csv')

# Abril
sinasc_abr = pd.read_csv('./input/SINASC_RO_2019_ABR.csv')

# Maio
sinasc_mai = pd.read_csv('./input/SINASC_RO_2019_MAI.csv')

# Junho
sinasc_jun = pd.read_csv('./input/SINASC_RO_2019_JUN.csv')

# Dezembro
sinasc_dez = pd.read_csv('./input/SINASC_RO_2019_DEZ.csv')
```

```python
# Convertendo a coluna 'DTNASC' para o formato de data e extraindo o mês

def preprocess_dataframe(df):
    df['DTNASC'] = pd.to_datetime(df['DTNASC'])
    df['MES'] = df['DTNASC'].dt.month
    return df
```

```python
# Aplicando a função de pré-processamento para todos os dataframes

sinasc_mar = preprocess_dataframe(sinasc_mar)
sinasc_abr = preprocess_dataframe(sinasc_abr)
sinasc_mai = preprocess_dataframe(sinasc_mai)
sinasc_jun = preprocess_dataframe(sinasc_jun)
sinasc_dez = preprocess_dataframe(sinasc_dez)
```

```python
# Função para gerar gráficos

def plot_graphs(df, mes):

    # Criar diretório, se necessário
    directory = f'./output/figs/{mes}/'
    if not os.path.exists(directory):
        os.makedirs(directory)

    plt.figure(figsize=(14, 7))

    # Média da idade da mãe por mês
    df.groupby('MES')['IDADEMAE'].mean().plot(kind='bar')
    plt.title(f'Média da Idade da Mãe em {mes}')
    plt.ylabel('Idade Média')
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.savefig(f'./output/figs/{mes}/media_idade_mae.png')
    plt.show()

    # Média do peso do bebê por mês
    df.groupby('MES')['PESO'].mean().plot(kind='bar', color='green')
    plt.title(f'Média do Peso dos Bebês em {mes}')
    plt.ylabel('Peso Médio (g)')
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.savefig(f'./output/figs/{mes}/media_peso_bebes.png')
    plt.show()

    # Média do APGAR1 por mês
    df.groupby('MES')['APGAR1'].mean().plot(kind='bar', color='blue')
    plt.title(f'Média do APGAR1 em {mes}')
    plt.ylabel('APGAR1 Médio')
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.savefig(f'./output/figs/{mes}/media_apgar1.png')
    plt.show()

    # Quantidade de nascimentos por tipo de gestação
    df.groupby('GESTACAO').size().plot(kind='bar', color='purple')
    plt.title(f'Quantidade de Nascimentos por Tipo de Gestação em {mes}')
    plt.ylabel('Quantidade')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.savefig(f'./output/figs/{mes}/qtd_nascimentos_gestacao.png')
    plt.show()

    # Quantidade de nascimentos por escolaridade da mãe
    df.groupby('ESCMAE').size().plot(kind='bar', color='orange')
    plt.title(f'Quantidade de Nascimentos por Escolaridade da Mãe em {mes}')
    plt.ylabel('Quantidade')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.savefig(f'./output/figs/{mes}/qtd_nascimentos_escolaridade_mae.png')
    plt.show()

```

```python
# Agora, gerando os gráficos para cada mês:

meses = {
    "Março": sinasc_mar,
    "Abril": sinasc_abr,
    "Maio": sinasc_mai,
    "Junho": sinasc_jun,
    "Dezembro": sinasc_dez
}

for mes, df in meses.items():
    plot_graphs(df, mes)
```
