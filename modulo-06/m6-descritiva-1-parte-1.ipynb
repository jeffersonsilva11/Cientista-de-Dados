#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Caminho para o arquivo CSV
caminho_arquivo = "sinasc_RO_2019.csv"

# Carregando o arquivo CSV
df_sinasc = pd.read_csv(caminho_arquivo)

# Visualizando as primeiras linhas do dataframe
df_sinasc.head()
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
# Variável APGAR1
df_sinasc['APGAR1'].value_counts().sort_index().plot(kind='bar')
plt.title('Distribuição da variável APGAR1')
plt.xlabel('APGAR1')
plt.ylabel('Frequência')
plt.show()

# Variável APGAR5
df_sinasc['APGAR5'].value_counts().sort_index().plot(kind='bar')
plt.title('Distribuição da variável APGAR5')
plt.xlabel('APGAR5')
plt.ylabel('Frequência')
plt.show()

# Variável SEXO
df_sinasc['SEXO'].value_counts().sort_index().plot(kind='bar')
plt.title('Distribuição da variável SEXO')
plt.xlabel('SEXO')
plt.ylabel('Frequência')
plt.show()

# Variável GESTACAO
df_sinasc['GESTACAO'].value_counts().sort_index().plot(kind='bar')
plt.title('Distribuição da variável GESTACAO')
plt.xlabel('GESTACAO')
plt.ylabel('Frequência')
plt.show()

# Variável ESTCIVMAE
df_sinasc['ESTCIVMAE'].value_counts().sort_index().plot(kind='bar')
plt.title('Distribuição da variável ESTCIVMAE')
plt.xlabel('ESTCIVMAE')
plt.ylabel('Frequência')
plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
# Agrupamento dos dados por 'CONSULTAS' e calculando a média do 'APGAR5'
df_grouped = df_sinasc.groupby('CONSULTAS')['APGAR5'].mean()
```

```python
# Criação de um gráfico de barras a partir dos resultados com o método plot do pandas.

df_grouped.plot(kind='bar')
plt.title('APGAR5 médio por CONSULTAS')
plt.xlabel('CONSULTAS')
plt.ylabel('APGAR5 médio')
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Gráfico de barras para a variável SEXO
df_sinasc['SEXO'].value_counts().plot(kind='bar', ax=axs[0, 0])
axs[0, 0].set_title('SEXO')

# Gráfico de barras para a variável GESTACAO
df_sinasc['GESTACAO'].value_counts().plot(kind='bar', ax=axs[0, 1])
axs[0, 1].set_title('GESTACAO')

# Gráfico de barras para a variável ESTCIVMAE
df_sinasc['ESTCIVMAE'].value_counts().plot(kind='bar', ax=axs[1, 0])
axs[1, 0].set_title('ESTCIVMAE')

# Gráfico de barras para a variável CONSULTAS
df_sinasc['CONSULTAS'].value_counts().plot(kind='bar', ax=axs[1, 1])
axs[1, 1].set_title('CONSULTAS')

plt.tight_layout()
plt.show()
```
