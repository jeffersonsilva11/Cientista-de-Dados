1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Leitura da base de dados
df = pd.read_csv("previsao_de_renda.csv")
````

````python
# Conversão da coluna de data para o formato correto
df['data_ref'] = pd.to_datetime(df['data_ref'])
````

````python
#Construção dos gráficos de barras para a variável 'sexo'
plt.figure(figsize=(12, 7))
ax = sns.countplot(x='data_ref', hue='sexo', data=df)

tick_labs = df['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()
ticks = ax.set_xticks(list(range(df['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation=45)

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title('Distribuição de Sexo ao Longo do Tempo')
plt.show()
````

````python
#Construção dos gráficos de barras para a variável 'posse_de_veículo'
plt.figure(figsize=(12, 7))
ax = sns.countplot(x='data_ref', hue='posse_de_veiculo', data=df)

tick_labs = df['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()
ticks = ax.set_xticks(list(range(df['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation=45)

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title('Distribuição de Posse de Veículo ao Longo do Tempo')
plt.show()
````

````python
#Construção dos gráficos de barras para a variável 'posse_de_imovel'
plt.figure(figsize=(12, 7))
ax = sns.countplot(x='data_ref', hue='posse_de_imovel', data=df)

tick_labs = df['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()
ticks = ax.set_xticks(list(range(df['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation=45)

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title('Distribuição de Posse de Imóvel ao Longo do Tempo')
plt.show()
````

2) Para cada variável qualitativa, construa um gráfico mostrando as médias de cada nível da variável no tempo e avalie a estabilidade. Houve alguma alteração importante? Você acha que se pode utilizar essas variáveis em uma modelagem?

````python
# Transforma a coluna "sexo" em uma variável binária
df['sexo'] = df['sexo'].map({'M': 0, 'F': 1})
````

````python
# Cria uma tabela de frequências
tab = pd.crosstab(df['data_ref'], df['sexo'])

# Calcula a proporção
tab_prop = tab.div(tab.sum(axis=1), axis=0)

plt.figure(figsize=(12, 7))
ax = tab_prop.plot.bar(stacked=True, figsize=(12, 7))
plt.title('Proporção de Sexo ao Longo do Tempo')
plt.ylabel('Proporção')
plt.xlabel('Data')
tick_labs = df['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()
ticks = ax.set_xticks(list(range(df['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation=45)
plt.legend(title="Sexo", labels=['Masculino', 'Feminino'], loc='upper left')
plt.show()
````

### Não acredito que nos gráficos acima exista uma mudança significativa que mereça ser investigada, visto que as alterações seguem com baixa diferença uma da outra.
