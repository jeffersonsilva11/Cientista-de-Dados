## Nmero de grupos

Nesta atividade vamos adotar uma abordagem bem pragmtica e avaliar agrupamentos hierrquicos com 3 e 4 grupos, por estarem bem alinhados com uma expectativa e estratgia do diretor da empresa. 

*Ateno*: Cuidado se quiser fazer o dendrograma, pois com muitas observaes ele pode ser mais complicado de fazer, e dependendo de como for o comando, ele pode travar o *kernell* do seu python.

```python
# Identificando as colunas no numricas
non_numeric_columns = df.select_dtypes(include=['object']).columns
print(non_numeric_columns)
```

```python
# Convertendo a coluna 'Month' para valores numricos
month_mapping = {
    'Feb': 2, 'Mar': 3, 'May': 5, 'June': 6, 'Jul': 7, 'Aug': 8, 'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12
}

df['Month'] = df['Month'].map(month_mapping)

# Aplicando codificao one-hot na coluna 'VisitorType'
df = pd.get_dummies(df, columns=['VisitorType'], drop_first=True)

# Normalizando os dados
scaler = StandardScaler()
df_normalized = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
```

```python
# Escolhendo o mtodo do cotovelo para ajudar na deciso, mesmo decidindo a usar 3 e 4
wcss = []

# Testando para at 10 clusters
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(df_normalized)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10,5))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Mtodo do Cotovelo')
plt.xlabel('Nmero de clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Execuo do K-means para 3
kmeans3 = KMeans(n_clusters=3, init='k-means++', random_state=42)
clusters3 = kmeans3.fit_predict(df_normalized)

# Execuo do K-means para 4
kmeans4 = KMeans(n_clusters=4, init='k-means++', random_state=42)
clusters4 = kmeans4.fit_predict(df_normalized)
```

```python
# Avaliando o coeficiente de silhueta
score3 = silhouette_score(df_normalized, clusters3)
score4 = silhouette_score(df_normalized, clusters4)

print(f"Coeficiente de Silhueta para 3 clusters: {score3}")
print(f"Coeficiente de Silhueta para 4 clusters: {score4}")
```

Pelo Mtodo do Cotovelo, parece que 3 ou 4 clusters so o ponto de "cotovelo", onde adicionar mais clusters no resulta em uma reduo significativa da WCSS.
O Coeficiente de Silhueta mostra uma leve melhoria quando passamos de 3 para 4 clusters.#### 

## Avaliao dos grupos

Construa os agrupamentos com a tcnica adequada que vimos em aula. No se esquea de tratar variveis qualitativas, padronizar escalas das quantitativas, tratar valores faltantes e utilizar a distncia correta.

Faa uma anlise descritiva para pelo menos duas solues de agrupamentos (duas quantidades diferentes de grupos) sugeridas no item anterior, utilizando as variveis que esto no escopo do agrupamento.
- Com base nesta anlise e nas anlises anteriores, decida pelo agrupamento final. 
- Se puder, sugira nomes para os grupos.

```python
# Definindo as colunas
variaveis_quantitativas = [
    "Administrative", "Administrative_Duration", "Informational", 
    "Informational_Duration", "ProductRelated", "ProductRelated_Duration",
    "BounceRates", "ExitRates", "PageValues", "SpecialDay"
]

variaveis_qualitativas = ["OperatingSystems", "Browser", "Region", "TrafficType"]

# Criando o pre-processador
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), variaveis_quantitativas),
        ('cat', OneHotEncoder(), variaveis_qualitativas)
    ])

# Criando o pipeline
pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                           ('cluster', KMeans(n_clusters=3))])
pipeline.fit(df)
```

```python
# Anlise descritiva dos grupos

df['cluster_3'] = pipeline.predict(df)
df.groupby('cluster_3').mean()
```
