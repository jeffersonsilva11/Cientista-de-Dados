```python
import pandas as pd
import matplotlib.pyplot as plt

# Caminho para o arquivo CSV
caminho_arquivo = "sinasc_RO_2019.csv"

# Carregando o arquivo CSV
df_sinasc = pd.read_csv(caminho_arquivo)

# Visualizando as primeiras linhas do dataframe
df_sinasc.head()
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
# Criar a figura
plt.figure(figsize=(8,6))

# Adicionar o boxplot à figura
df_sinasc['IDADEMAE'].plot(kind='box')

# Mostrar o título
plt.title('Boxplot da Idade da Mãe')

# Mostrar o gráfico
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
# Criar a figura
plt.figure(figsize=(8,6))

# Adicionar o boxplot à figura
df_sinasc['PESO'].plot(kind='box')

# Mostrar o título
plt.title('Boxplot do Peso do Bebê')

# Mostrar o gráfico
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
# Criar a figura
plt.figure(figsize=(8,6))

# Adicionar o histograma à figura
df_sinasc['IDADEMAE'].plot(kind='hist', bins=30, rwidth=0.8)

# Mostrar o título
plt.title('Histograma da Idade da Mãe')

# Mostrar o gráfico
plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
# Criar a figura
plt.figure(figsize=(8,6))

# Adicionar o histograma à figura
df_sinasc['PESO'].plot(kind='hist', bins=30, rwidth=0.8)

# Mostrar o título
plt.title('Histograma do Peso do Bebê')

# Mostrar o gráfico
plt.show()
```
