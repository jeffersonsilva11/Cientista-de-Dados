Neste exercício vamos trabalhar com a base "Heart Disease Data Set" disponível no site do [UCI](https://archive.ics.uci.edu/ml/datasets/heart+disease).

Temos 303 indivíduos e vamos utilizar 13 variáveis disponíveis para prever a presença de doença cardíaca na variável ```num```(a 14$^a$ variável).

A descrição das variáveis está recortada abaixo:
- age: idade do paciente em anos
- sex: sexo (1 = male; 0 = female)  
- cp: tipo de dor no peito
  - 1: angina típica
  - 2: angina atípica
  - 3: dor não-angina
  - 4: assintomático
- trestbps: pressão sanguínea em repouso (em mm Hg na admissão ao hospital
- chol: colesterol sérico em mg/dl
- fbs: (açúcar no sangue em jejum > 120 mg/dl) (1 = True; 0 = False)
- restecg: resultados eletrocardiográficos em repouso
  - 0: normal
  - 1: tendo anormalidade da onda ST-T (Inversões de onda T e / ou ST com elevação ou depressão de > 0.05 mV)
  - 2: mostrando hipertrofia ventricular esquerda provável ou definitiva pelos critérios de Estes
- thalach: frequência cardíaca máxima alcançada
- exang: angina induzida por exercício(1 = sim; 0 = não)
- oldpeak = Depressão de ST induzida por exercício em relação ao repouso
- slope: Depressão de ST induzida por exercício em relação ao repouso
  - 1: inclinação ascendente
  - 2: estável
  - 3: inclinação descendente
- ca: número de vasos principais (0-3) coloridos por fluorosopia
- thal: 3 = normal; 6 = defeito corrigido; 7 = defeito reversível
- num: diagnóstico de doença cardíaga (status de doença angiográfica)

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.formula.api as smf
````

````python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data'

df = pd.read_csv(url, 
                 names=['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg',
                        'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'num'])
df.head()
````

````python
df['flag_doente'] = (df['num'] != 0).astype('int64')
df
````

1. Monte uma bivariada para a variável ```sex``` com as seguintes estatísticas:

- Número de *saudáveis*
- Número de *doentes*
- Total
- Quantidade média de doentes
- *Odds*
- *Odds ratio* (Use o total como referência)
- Logito (LOG(*Odds*))
- O LOG do *Odds ratio* (com o total como referência) é chamado de *Weight of Evidence* (WOE). Calcule-o.

````python
# 1. Número de saudáveis
saudaveis = df[df['flag_doente'] == 0].groupby('sex').size()

# 2. Número de doentes
doentes = df[df['flag_doente'] == 1].groupby('sex').size()

# 3. Total
total = df.groupby('sex').size()

# 4. Quantidade média de doentes
media_doentes = df.groupby('sex')['flag_doente'].mean()

# 5. Odds (probabilidade de ser doente / probabilidade de ser saudável)
odds = media_doentes / (1 - media_doentes)

# 6. Odds ratio
odds_ratio_masculino = odds[1] / odds[0]
odds_ratio_feminino = odds[0] / odds[1]

# 7. Logito (LOG(Odds))
logito = np.log(odds)

# 8. Weight of Evidence (WOE)
woe_masculino = np.log(odds_ratio_masculino)
woe_feminino = np.log(odds_ratio_feminino)

# Criando um dataframe para visualização
bivariada = pd.DataFrame({
    'Saudáveis': saudaveis,
    'Doentes': doentes,
    'Total': total,
    'Média Doentes': media_doentes,
    'Odds': odds,
    'Logito': logito,
    'WOE Masculino': woe_masculino,
    'WOE Feminino': woe_feminino
})

print(bivariada)
````

## Com base nesta análise, ser do sexo masculino parece aumentar o risco de ter doença cardíaca em comparação com o feminino

2. Monte uma bivariada para a variável ```age``` com as seguintes estatísticas:
- Categorize a variável em 5 grupos
- Calcule as mesmas quantidades calculadas no execício anterior
- Calcule também o valor médio da variável ```age```

````python
# 1. Categorizando da idade
df['age_cat'] = pd.cut(df['age'], bins=5, labels=['Very Young', 'Young', 'Middle Age', 'Senior', 'Very Senior'])

# 2. Calculando estatísticas bivariadas
biv_age = df.groupby('age_cat').agg({
    'flag_doente': ['sum', 'count', 'mean'],
    'age': 'mean'
}).reset_index()

biv_age.columns = ['Age Category', 'Doentes', 'Total', 'Média Doentes', 'Média Age']
biv_age['Saudáveis'] = biv_age['Total'] - biv_age['Doentes']
biv_age['Odds'] = biv_age['Doentes'] / biv_age['Saudáveis']
biv_age['Logito'] = np.log(biv_age['Odds'])

# Cálculo do WOE usando o total como referência
total_odds = biv_age['Doentes'].sum() / biv_age['Saudáveis'].sum()
biv_age['WOE'] = np.log(biv_age['Odds'] / total_odds)

# 3. Valor médio da variável age (já calculado acima na coluna 'Média Age')

# Printando os resultados:
print(biv_age)
````

### De forma resumida, podemos observar que a prevalência da doença cardíaca tende a ser maior entre os indivíduos "Senior". O WOE (Weight of Evidence) indica que o grupo "Senior" tem maior evidência (valor positivo de WOE) em relação ao total de ter doença cardíaca, enquanto os outros grupos (especialmente "Very Young" e "Young") têm menor evidência (valores negativos de WOE).

3. Visualize a informação da bivariada construída acima através de um gráfico da sua escolha.

````python
# Criando figura e eixo
fig, ax1 = plt.subplots(figsize=(10, 6))

# Criando o gráfico de barras para a Média de Doentes
sns.barplot(x='Age Category', y='Média Doentes', data=biv_age, ax=ax1, palette='viridis')

# Criando um eixo secundário para a Média da Idade
ax2 = ax1.twinx()
sns.lineplot(x='Age Category', y='Média Age', data=biv_age, ax=ax2, color='red', marker='o', label='Média Age')

# Definindo títulos e labels
ax1.set_title('Média de Doentes por Categoria de Idade e Média de Idade', fontsize=15)
ax1.set_xlabel('Categoria de Idade', fontsize=14)
ax1.set_ylabel('Média de Doentes', fontsize=14)
ax2.set_ylabel('Média da Idade', fontsize=14)
ax2.legend(loc='upper left')

plt.tight_layout()
plt.show()
````
