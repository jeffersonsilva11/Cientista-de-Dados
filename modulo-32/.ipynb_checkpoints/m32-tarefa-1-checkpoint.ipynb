# Probabilidade - Tarefa

```python
import numpy as np
from scipy import stats
```

1. Uma máquina embala ração para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatório, com distribuição Normal, com média 15kg, mas com desvio padrão 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produção tenha mais que 15,2 kg.

```python
# Definindo a média e o desvio padrão
mu = 15.0
sigma = 0.1

# Obtenhendo a probabilidade de que um pacote tenha peso menor ou igual a 15,2 kg
prob_menor_ou_igual_15_2 = stats.norm.cdf(15.2, mu, sigma)

# Probabilidade de que um pacote tenha mais que 15,2 kg
prob_maior_15_2 = 1 - prob_menor_ou_igual_15_2

print(f"A probabilidade de um pacote ter mais que 15,2 kg é {prob_maior_15_2:.4f}")
```

2. Da máquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

```python
# Definindo a média e o desvio padrão
mu = 15.0
sigma = 0.1

# Obtendo o percentil de ordem 95%
percentil_95 = stats.norm.ppf(0.95, mu, sigma)

print(f"O percentil de ordem 95% para o peso dos pacotes é {percentil_95:.4f} kg")
```

3. Suponha duas variáveis aleatórias $Y_1$ e $Y_2$, a primeira com distribuição Normal Padrão (média=0, variância=1), a segunda com distribuição *t-studdent* com 5 graus de liberdade (também padrão). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na célula abaixo:

```python
# Definindo os valores em Y
Y = np.linspace(-6, 6, 200)

# Calculando a probabilidade de Y1 ser menor que cada valor em Y usando a CDF da distribuição normal padrão
prob_Y1 = stats.norm.cdf(Y, 0, 1)

# Calculando a probabilidade de Y2 ser menor que cada valor em Y usando a CDF da distribuição t-student com 5 graus de liberdade
prob_Y2 = stats.t.cdf(Y, 5)

# Imprimindo algumas das probabilidades para verificação
for y, p1, p2 in zip(Y[::40], prob_Y1[::40], prob_Y2[::40]):  # Pega a cada 40 valores para não imprimir demais
    print(f"Para Y = {y:.2f}, P(Y1 < Y) = {p1:.4f}, P(Y2 < Y) = {p2:.4f}")

# prob_Y1 e prob_Y2 contêm as probabilidades correspondentes para cada valor em Y
```

4. Faça uma visualização gráfica dos dados do item anterior.

```python
import matplotlib.pyplot as plt

# Plotando as probabilidades
plt.figure(figsize=(10, 6))

# Gráfico para a variável Y1 com distribuição Normal
plt.plot(Y, prob_Y1, label='Normal', color='blue')

# Gráfico para a variável Y2 com distribuição t-student
plt.plot(Y, prob_Y2, label='t(5)', color='red')

# Detalhes do gráfico
plt.title('CDF das variáveis aleatórias Y1 e Y2')
plt.xlabel('Valor de Y')
plt.ylabel('Probabilidade')
plt.legend()
plt.grid(True)
plt.show()
```

5. 3. Suponha duas variáveis aleatórias $Y_1$ e $Y_2$, a primeira com distribuição Normal Padrão (média=0, variância=1), a segunda com distribuição *t-studdent* com 5 graus de liberdade (também padrão). Calcule, para cada uma, o percentil (a inversa da *CDF*) para cada um dos valores do objeto *P* abaixo:

```python
Y=np.linspace(0, 1, 20)

from scipy.stats import norm, t

P = np.linspace(0, 1, 20)

# Para a distribuição Normal
quantiles_Y1 = norm.ppf(P)

# Para a distribuição t-student com 5 graus de liberdade
quantiles_Y2 = t.ppf(P, df=5)

for p, q1, q2 in zip(P, quantiles_Y1, quantiles_Y2):
    print(f"Para P = {p:.2f}, Y1: {q1:.4f}, Y2: {q2:.4f}")
```

6. Faça um gráfico com os dados do item anterior.

```python
plt.figure(figsize=(10, 6))

# Plotando os dados
plt.plot(P, quantiles_Y1, label='Normal', color='blue')
plt.plot(P, quantiles_Y2, label='t(5)', color='red')

# Adicionando legendas e títulos
plt.title('Quantis das distribuições Y1 e Y2')
plt.xlabel('Probabilidade')
plt.ylabel('Valor do Quantil')
plt.legend()
plt.grid(True)

plt.show()
```
